```python
import cv2
import pandas as pd
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("map8.png")
# print(img)

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

ret,thresh = cv2.threshold(gray,70,255,0)

# cv2.imshow("Binary Image",thresh)
# cv2.waitKey(0)
# cv2.destroyAllWindows()

# print((thresh[300]))

data1 = pd.DataFrame(thresh)
data1 = data1.replace(0,1)
data1 = data1.replace(255,0)
data1.to_excel("output8.xlsx")

plt.imshow(data1, vmin=0, vmax=1, origin='lower')
```
